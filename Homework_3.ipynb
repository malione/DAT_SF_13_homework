### 1.	Use the file bank.csv to explore the dataset. <small>Observe the features: Are they numbers? Are they strings? Are they binary? Are they continuous?</small>

```python
# read in bank data
data = pd.read_csv('bank.csv', sep=';')
data.describe()
```

```python
data.head()
```

- Some of the features are numbers: age, balance, day, duration, campaign, pdays, previous
- The others are strings: job, marital, education, default, housing, loan, contact, month, poutcome, y

### 2.	Learn about label encoders at the following link and use what you learn to transform the features to numerical features. <small>http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html</small>

```python
numerical_features = data.get(['age', 'balance', 'day', 'duration', 'campaign','pdays','previous'])
text_features = data.get(['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y'])

months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
labels = [];
for col in text_features.columns:
    vallist = np.unique(data[col])
    if col == 'month':
        labels += [col[0] + '=' + i[:3] for i in months]
        intdict = dict(zip(months, range(1,13)))
    else:
        labels += [col[0] + '=' + i[:4] for i in vallist]
        intdict = dict(zip(vallist, range(len(vallist))))
    for k in intdict.keys():                 
        text_features[col].replace(k, intdict[k], inplace=True)

print len(labels)
print labels
text_features.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse=False)
hotcoded = enc.fit_transform(text_features).astype(float)
print hotcoded.shape
coded_features = pd.DataFrame(hotcoded, index=range(hotcoded.shape[0]), columns=labels)
coded_features.head()
```

### 3.	Build a decision tree model to predict whether a prospect will buy the product.

```python
newdata = np.array(pd.concat([numerical_features, coded_features], axis=1))
bank_X, bank_y = newdata[:, :-2], newdata[:, -1]

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(bank_X, bank_y, test_size=0.20, random_state=12)


from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3, min_samples_leaf=5)
clf = clf.fit(X_train,y_train)
```

### 4.	Evaluate the accuracy of your decision tree model using cross validation.

```python
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(clf, newdata[:, :-2], newdata[:, -1], cv=5)

# Print the resulting scores and their mean, which represents the overall accuracy.
print "Cross-validation scores:\t", scores
print "Overall classifier accuracy:\t", scores.mean()
```

### 5.	Repeat the analysis and cross validation with the file bank-additional-full.csv. <small>How does the performance of the model change (with the additional training examples and additional features)?</small>

```python
# read in additional full bank data
data = pd.read_csv('bank-additional-full.csv', sep=';')
print data.columns.values
data.describe()
```

```python
# Redefine the numerical features and text features
numerical_features = data.get(['age', 'duration', 'campaign', 'pdays','previous', 'emp.var.rate', 'cons.price.idx',
                               'cons.conf.idx', 'euribor3m', 'nr.employed'])
text_features = data.get(['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week',
                          'poutcome', 'y'])

labels = [];
for col in text_features.columns:
    vallist = np.unique(data[col])
    if col == 'month':
        labels += [col[0] + '=' + i[:3] for i in vallist]
        intdict = dict(zip(vallist, range(len(vallist))))
    elif col == 'education':
        labels += [col[0] + '=' + i[:9] for i in vallist]
        intdict = dict(zip(vallist, range(len(vallist))))        
    else:
        labels += [col[0] + '=' + i[:4] for i in vallist]
        intdict = dict(zip(vallist, range(len(vallist))))
    for k in intdict.keys():                 
        text_features[col].replace(k, intdict[k], inplace=True)

print len(labels)
print labels

# Use OneHotEncoder to create the sparse coded text values
enc = OneHotEncoder(sparse=False)
hotcoded = enc.fit_transform(text_features).astype(float)
print hotcoded.shape
coded_features = pd.DataFrame(hotcoded, index=range(hotcoded.shape[0]), columns=labels)
coded_features.head()
```

```python
# Build a decision tree model to predict whether a prospect will buy the product.
newdata = np.array(pd.concat([numerical_features, coded_features], axis=1))
bank_X, bank_y = newdata[:, :-2], newdata[:, -1]

X_train, X_test, y_train, y_test = train_test_split(bank_X, bank_y, test_size=0.20, random_state=12)
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3, min_samples_leaf=5)
clf = clf.fit(X_train,y_train)

# Evaluate the accuracy using cross validatation
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(clf, newdata[:, :-2], newdata[:, -1], cv=5)

# Print the resulting scores and their mean, which represents the overall accuracy.
print "Cross-validation scores:\t", scores
print "Overall classifier accuracy:\t", scores.mean()
```
