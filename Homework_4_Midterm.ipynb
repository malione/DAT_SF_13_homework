**Data Set Information:**

The "spam" concept is diverse: advertisements for products/web sites, make money fast schemes, chain letters, pornography...

**Attribute Information:**

The last column of 'spambase.data' denotes whether the e-mail was considered spam (1) or not (0), i.e. unsolicited commercial e-mail. Most of the attributes indicate whether a particular word or character was frequently occuring in the e-mail. The run-length attributes (55-57) measure the length of sequences of consecutive capital letters.

###Definitions of the attributes:

**48 continuous real [0,100] attributes of type word_freq_WORD**

= percentage of words in the e-mail that match WORD, i.e. 100 * (number of times the WORD appears in the e-mail) / total number of words in e-mail. A "word" in this case is any string of alphanumeric characters bounded by non-alphanumeric characters or end-of-string.

**6 continuous real [0,100] attributes of type char_freq_CHAR**

= percentage of characters in the e-mail that match CHAR, i.e. 100 * (number of CHAR occurences) / total characters in e-mail

**1 continuous real [1,...] attribute of type capital_run_length_average**

= average length of uninterrupted sequences of capital letters

**1 continuous integer [1,...] attribute of type capital_run_length_longest**

= length of longest uninterrupted sequence of capital letters

**1 continuous integer [1,...] attribute of type capital_run_length_total**

= sum of length of uninterrupted sequences of capital letters = total number of capital letters in the e-mail

**1 nominal {0,1} class attribute of type spam**

= denotes whether the e-mail was considered spam (1) or not (0), i.e. unsolicited commercial e-mail.

###Exercise 2: Build a simple logistic regression and visualize it
use the variable "capital_run_length_longest" to predict "is_spam" How accurate is this single feature?

```python
#your code here
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Use capital_run_length_longest as the only feature
target = spam_data.is_spam
features_array = pd.DataFrame(spam_data['capital_run_length_longest'])
# print target.shape
# print features_array.shape

# Train and then test on the entire data set
lr = LogisticRegression(C=1)
lr.fit(features_array, target)

target_predicted = lr.predict(features_array)
accuracy_score(target, target_predicted)
```

###Exercise 3: Use train-test spit to split your data at a 30% mark and run another logistic regression using all variables
use random state = 12 so that we can compare results

```python
#your code here
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

target = spam_data.is_spam
features_array = spam_data[spam_data.columns[0:-2]]
features_train, features_test, target_train, target_test = train_test_split(
                                                        features_array, target, test_size=0.30, random_state=12)
lr = LogisticRegression(C=1)
lr.fit(features_train, target_train)

target_predicted = lr.predict(features_test)
accuracy_score(target_test, target_predicted)
```

###Exercise 4: Apply cross validation to see how the model fares across different splits of your data
Use cross validation to score your model

```python
#your code here
from sklearn.cross_validation import cross_val_score

target = spam_data.is_spam
features_array = spam_data[spam_data.columns[0:-2]]
scores = cross_val_score(lr, features_array, target, cv=10)

# Print the resulting scores and their mean, which represents the overall accuracy.
print "Cross-validation scores:\t", scores[:len(scores)/2], "\n\t\t\t\t", scores[len(scores)/2:]
print "Overall classifier accuracy:\t", scores.mean()
```

###Exercise 5: Compare Performance of Logistic Regression to KNN with 3 neighbors
Which model is more accurate?

```python
#your code here
from sklearn import neighbors

clf = neighbors.KNeighborsClassifier(3, weights='uniform')
scores = cross_val_score(clf, features_array, target, cv=10)

# Print the resulting scores and their mean, which represents the overall accuracy.
print "Cross-validation scores:\t", scores[:len(scores)/2], "\n\t\t\t\t", scores[len(scores)/2:]
print "Overall classifier accuracy:\t", scores.mean()
```

* The Logistic Regression is slightly more accurate (average accuracy of 92%) than the KNN with 3 neighbors (average accuracy of 88%).

###Exercise 7: Plot the ROC Curve for the logistic regression you chose¶

```python
#your code here
from sklearn.metrics import roc_curve

# Produe y_prob for the entire dataset
X_test = features_array
y_test = target
y_prob = lr.predict_proba(X_test)

# Compute ROC curve the class
fpr, tpr, _ = roc_curve(y_test, y_prob[:,1])

# Plot of a ROC curve for a specific class
plt.figure()
plt.plot(fpr, tpr, label='ROC curve')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.legend(loc="lower right")
plt.show()
```

###Exercise 8: Demonstrate how the accurary of your predictions changes when you set your threshold to levels other than 50%

```python
#your code here

# Using the same LR classifier determined above, I'll use the probabality array to test different threshold levels.
y_true = target
y_prob = lr.predict_proba(X_test)

for th in np.arange(0.25, 0.75, .02):
    y_pred = 0.5 * (np.sign(y_prob[:,1] - th) + 1)
    print "Threshold: {:.2f}\t\tAccuracy: {}".format(th, accuracy_score(y_true, y_pred))
```

###Exercise 9: Discuss the pro's/con's of moving the threshold away from 50%, why is/isn't this a good idea?

####your essay-style answer here
Given the ROC curve for this logistic regression classifier, it makes sense to have the threshold be lower than 50%.

Looking at the ROC curve we see that it rises quite sharply early on, so the TPR can get really high as the FPR rises only slowly. Beyond an FPR of 0.15, the TPR only rises very little as the FPR rises considerably. It's not easy to discern by observation how the threshold level affects this, but my test in Exercise 8 shows a peak accuracy when the threshold is set to 0.43.

Based on this, I conclude that it is a good idea to move the threshold slightly below 50%. As it drops lower than 50%, the pro is that the incidence of false negatives will go down, but at the same time, the incidence of false positives goes up. The ROC tells us that we are more likely to lose false negatives than we are to gain false positives when we reduce the threshold down to 43%, so the overall accuracy improves. The only other mitigating factor would be the relative cost incurred for false negatives as opposed to false positives. If false positives are relatively costly to address, then we may wish to maintain a 50% threshold, or possibly even raise the threshold to reduce their incidence.
